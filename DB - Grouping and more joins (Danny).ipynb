## 1) Using a single join and a grouping, write a query that allows you to find the `forename`, `surname`, `driverid` and the total number of wins for each driver in the `f1db` schema.

```python
%%sql SELECT forename, surname, drivers.driverid, count(*) AS no_wins FROM drivers
JOIN results on results.driverid = drivers.driverid WHERE results.position = 1
GROUP BY forename, surname, drivers.driverid ORDER BY count(*) DESC;
```

## 2) Write a query that joins the tables results, constructors and drivers to show the amount of times a driver has driven a car from a constructor. (...) Your table should have three fields: `drivers.driverref`, `constructors.name` and `count(*)` (...)

```python
%%sql SELECT drivers.driverref, constructors.name, count(*) FROM drivers
JOIN results on results.driverid = drivers.driverid
JOIN constructors USING (constructorid)
GROUP BY drivers.driverref, constructors.name ORDER BY count(*) DESC;
```

## 3) Now we have found the most driven vehicles even for those who didn't finish the race. Extend the query from 2 by removing all tuples from the result where the status is not "Finished".

```python
%%sql SELECT drivers.driverref, constructors.name, count(*) FROM drivers
JOIN results on results.driverid = drivers.driverid
JOIN constructors USING (constructorid)
JOIN status USING (statusid) 
WHERE status='Finished'
GROUP BY drivers.driverref, constructors.name ORDER BY count(*) DESC
```

## 4) Create a third (and new) query that finds the amount of milliseconds spent in pitstop (see the `pitstops` relation) by each unique combinations of `driverid` and `raceid`.

```python
%%sql CREATE OR REPLACE VIEW ms_pitstops AS
SELECT drivers.driverid, pitstops.raceid, drivers.driverref, SUM(pitstops.milliseconds) AS pitstop_total_ms FROM drivers
JOIN pitstops on pitstops.driverid = drivers.driverid
GROUP BY drivers.driverid, pitstops.raceid
ORDER BY sum(pitstops.milliseconds) DESC;
```

```python
%sql SELECT * FROM ms_pitstops LIMIT 50;
```

## 5) Time to put everything together. Use the query for 4 as a join subquery in query 3 (see slides on "a query within a query") to find the total amount of pitstop time, for each result in query 3

```python
%%sql SELECT drivers.driverref, constructors.name, count(*), SUM(pitstop_total_ms) AS pitstop_total_ms 
FROM drivers
JOIN results on results.driverid = drivers.driverid
JOIN constructors USING (constructorid)
JOIN status USING (statusid)
INNER JOIN (SELECT * FROM ms_pitstops) AS pitstop_time USING (driverref, raceid)
WHERE status='Finished'
GROUP BY drivers.driverref, constructors.name
ORDER BY count(*) DESC;
```

## 6) As a last thing we want to find the driver with the least pistop time. However we want to control for the amount of races (see explanation below), so we need to include one last column: the average pitstop time per race. Your table should be sorted based on the ascending values of that average pitstop time.

```python
%%sql SELECT drivers.driverref, constructors.name, SUM(pitstop_total_ms) AS pitstop_total_ms, AVG(pitstop_total_ms) AS pitstop_avg_ms
FROM drivers
JOIN results on results.driverid = drivers.driverid
JOIN constructors USING (constructorid)
JOIN status USING (statusid)
JOIN races USING (raceid)
INNER JOIN (SELECT * FROM ms_pitstops) as pitstop_time USING (driverref, raceid)
WHERE status='Finished'
GROUP BY drivers.driverref, constructors.name
ORDER BY AVG(pitstop_total_ms) ASC;
```
